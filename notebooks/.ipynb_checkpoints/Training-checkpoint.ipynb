# Training model from cleaned dataset:

## Spliting data into Training and Testing Set:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Data.csv')
df.head()
```

```python
x = df.drop('charges', axis = 1)
y = df['charges']
```

```python
import numpy as np
y_log = np.log(y) #Creating long to avoid negetive charges
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_log, test_size = 0.20, random_state = 42)
```

* We made x as features except charges, and y as just charges.
* Then we made x_train, x_test, y_train, y_test, where x_train is taking features from training except charges, x_test is fearures from test except charges, y_train is charges from training set and y_test is charges from testing data.
* Now we use both x_train and y_train to train our model.
* But we use x_test to check the performance of model and the output we got, we can call it y_predict.
* Then we can compare y_test and y_predict to check the accuracy of the model.

# Creating model:

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## Testing model using testing dataset

```python
y_pred_log = model.predict(x_test)
y_pred = np.exp(y_pred_log)
```

```python
y_pred
```

```python
y_actual_test = np.exp(y_test)
y_actual_test
```

# Testing model:

## Evaluting model performace:

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_actual_test, y_pred)

n = x_test.shape[0]
p = x_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
```
